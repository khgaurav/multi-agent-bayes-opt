```python
import os
import numpy as np
from numpy import load
import matplotlib.pyplot as plt
```

```python
# for denormalizing time alloc
lb = 0.1
ub = 1.9
lb_i = np.ones(4)*lb
ub_i = np.ones(4)*ub
```

```python
# Load drone 1 time allocations
with open("traj_13_optimize_alpha_scaled_init_dataset_full.npy", "rb") as f:
    X_L = np.load(f)
    Y_L = np.load(f)

# lb and ub scaling to denorm X_L
lb_i = np.ones((len(X_L),4))*lb
ub_i = np.ones((len(X_L),4))*ub
X_L = lb_i + X_L*(ub_i-lb_i)

# segment labels
labels = ['Segment 1','Segment 2','Segment 3','Segment 4']

# plot histograms for each segment
plt.figure(figsize=(16, 4)) 
for i in range(4):
    plt.subplot(1, 4, i + 1) 
    plt.hist(X_L[:, i], bins=20, alpha=0.75, color=f'C{i}')
    plt.title(labels[i])
    plt.xlabel('Time Allocation')
    plt.ylabel('Frequency')

# Adjust layout to prevent overlap
plt.suptitle('Drone 1 trajectory segment sample distributions', fontsize=16)
plt.tight_layout()
plt.show()
```
